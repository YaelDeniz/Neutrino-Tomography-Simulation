```python
#/home/dehy0499/NuOscillation-Tomography/Neutrino-Tomography-Simulation/SimulationResults/PreliminaryResults/IntEvents/Events_Simulation_pancake_100Mton_1-20GeV_128-171Zen_-50-50Az

import numpy as np
import matplotlib.pylab as plt
import csv
import matplotlib.pyplot as plt
import sys
import math
from matplotlib.colors import LogNorm 

np.float = float  # Redirect any 'np.float' usage to 'float'

def csv_to_matrix(file_path):
    # Load CSV data into a numpy matrix
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        data = np.array([list(map(float, row)) for row in reader])

    return data

resultPath = "/home/dehy0499/NuOscillation-Tomography/Neutrino-Tomography-Simulation/SimulationResults/PreliminaryResults/MatterExample/"
SimulationPath = "Events_MSWExamples_100Mton_1-10GeV_90-180Zen_0-360Az/"



ligthFile   = "MSWExamples_prem_light_1_100Zen0Az100Enu_0.csv"
ligthEventFile  = resultPath+SimulationPath+ligthFile
ligthEvents  = csv_to_matrix(ligthEventFile)

heavyFile   = "MSWExamples_prem_heavy_1_100Zen0Az100Enu_0.csv"
heavyEventFile  = resultPath+SimulationPath+heavyFile
heavyEvents  = csv_to_matrix(heavyEventFile)

premFile   = "MSWExamples_prem_44layers_1_100Zen0Az100Enu_0.csv"
premEventFile  = resultPath+SimulationPath+premFile
premEvents  = csv_to_matrix(premEventFile)
    
cosZenith = np.unique(ligthEvents[:,0])
ene = np.unique(ligthEvents[:,1])
cthv, enev = np.meshgrid(cosZenith, ene)
    
NintVac = ligthEvents[:,2] # PREM Earth
NintVacmesh = NintVac.reshape(len(cosZenith),len(ene), order = "F")
SumNoEvac = NintVacmesh.sum(axis=0)

NintLight = ligthEvents[:,3] # Alternative Earth Model
NintLightmesh = NintLight.reshape(len(cosZenith),len(ene), order = "F")
SumNoELight = NintLightmesh.sum(axis=0)

NintHeavy = heavyEvents[:,3] # Alternative Earth Model
NintHeavymesh = NintHeavy.reshape(len(cosZenith),len(ene), order = "F")
SumNoHeavy = NintHeavymesh.sum(axis=0)

NintPrem = premEvents[:,3] # Alternative Earth Model
NintPremmesh = NintPrem.reshape(len(cosZenith),len(ene), order = "F")
SumNoPrem = NintPremmesh.sum(axis=0)


plt.plot(cosZenith, SumNoEvac, color = 'r', linestyle = '-', 
         marker = '.') 
plt.plot(cosZenith, SumNoELight, color = 'b', linestyle = '-', 
         marker = '.') 
plt.plot(cosZenith, SumNoHeavy, color = 'g', linestyle = '-', 
         marker = '.') 
plt.plot(cosZenith, SumNoPrem, color = 'k', linestyle = '-', 
         marker = '.') 




```

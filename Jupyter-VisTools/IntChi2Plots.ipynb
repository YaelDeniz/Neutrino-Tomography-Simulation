```python
import numpy as np
import matplotlib.pylab as plt
import csv
import matplotlib.pyplot as plt
import sys
import math
from matplotlib.colors import LogNorm 

np.float = float  # Redirect any 'np.float' usage to 'float'

def csv_to_matrix(file_path):
    # Load CSV data into a numpy matrix
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        data = np.array([list(map(float, row)) for row in reader])

    return data

plt.rc('font', family='serif')
```

```python
# Load the CSV file
result_folder = "/home/dehy0499/NuOscillation-Tomography/Neutrino-Tomography-Simulation/SimulationResults/PreliminaryResults/IntChi2/"
sim_folder = "IntChi2_-density-2-th-pancake1000_100_1-10_128-171_-50-50_2024-12-04/"
chi2rho_nue = result_folder +sim_folder+ "IntChi2_cth_-density-2-th-pancake1000_100nu0_Int60-100-60.csv"  # Replace with your file path
chi2rho_numu = result_folder +sim_folder+ "IntChi2_cth_-density-2-th-pancake1000_100nu1_Int60-100-60.csv"

chi2e = csv_to_matrix(chi2rho_nue)
chi2mu = csv_to_matrix(chi2rho_numu)

ne= chi2mu[:,2]
chi2 = chi2mu[:,4] + chi2e[:,4]


  
plt.plot(ne, chi2, color = '#156082', linestyle = '-', 
         marker = 'o') 
plt.plot(ne, 3*chi2, color = '#156082', linestyle = '-.', 
         marker = 'o') 
plt.plot(ne, 5*chi2, color = '#156082', linestyle = ':', 
         marker = 'o') 

# specifying Gaussian CL
plt.axhline(y = 1, color = '#7F7F7F', linestyle = '--') 
plt.axhline(y = 4, color = '#7F7F7F', linestyle = '--') 
plt.axhline(y = 9, color = '#7F7F7F', linestyle = '--') 

plt.xticks(rotation = 25) 
plt.xlabel(r'$n_e$') 
plt.ylabel(r'$\Delta \chi^2$') 
plt.title(r'Observed sensitivity to $n_e$ [Slab]', fontsize = 15) 
plt.legend(["100 Mton-year","300 Mton-year","500 Mton-year"])
plt.grid()  
plt.savefig('ObsRho2Slab.png')
plt.show() 
```
